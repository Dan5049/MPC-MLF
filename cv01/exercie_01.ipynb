# Exercise 1.5-1:

The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13. The next number in this series above is 8 + 13 = 21



Create a function that takes an integer as an input and return a list that contains all fibonaci numbers smaller than input integer

example:

*print(fibonaci(10))*

[0, 1, 1, 2, 3, 5, 8]


```python
def fibonaci(upper_threshold: int) -> list:
  fib = [0, 1]

  while upper_threshold > fib[-1] + fib[-2]:
    fib.append(fib[-2] + fib[-1])
  
  return [fib]

print(fibonaci(10))
```

```python
import numpy as np
import time

arr = np.arange(25) + 1
arr = arr[::-1]
arr = arr.reshape(5, 5)
limit = 5
#print(arr)

def arr_limiter1(arr_in1: np, limit1: int):
    lim_arr1 = arr_in1
    for i in range(len(arr_in1)):
        for j in range(len(arr_in1[0])):
            if arr_in1[i, j] > limit1:
                arr_in1[i, j] = 0 
    return [lim_arr1]

def arr_limiter2(arr_in2: np, limit2: int):
    lim_arr2 = arr_in2
    lim_arr2[lim_arr2 > limit2] = 0
    return [lim_arr2]

tim1 = time.perf_counter()
arr_limiter1(arr, limit)
tim2 = time.perf_counter()
arr_limiter2(arr, limit)
tim3 = time.perf_counter()

print("Time with loops:    ", tim2 - tim1)
print("Time without loops: ", tim3 - tim2)

print(arr_limiter1(arr, limit))
print(arr_limiter2(arr, limit))
```
